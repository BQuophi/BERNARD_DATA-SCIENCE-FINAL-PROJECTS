```python
# We begin by importing the necessary python libraries we will need for gathering the data
from bs4 import BeautifulSoup as bs
import pandas as pd
import requests as rq
```

```python
url ="https://www.yarnplaza.com/product/4216/dmc-natura-xl.html" # Website consisting of info for Natura XL for the DMC Brand
res = rq.get(url)
content = res.text
print(content) # Getting content from the url
```

```python
page = bs(content, 'lxml')
_page = bs(content, 'html.parser')
```

```python
# Now, lets begin by getting the name of the natura_xl product from the DMC brand
xl_name = page.find_all("span", class_="variants-title-txt")
# We now get the name of the product.Notice that we are using a list comprehension together with string methods to get the data we need.
natura_xl_name = [item.text.strip() for item in xl_name] 
print(natura_xl_name)
```

```python
# NEXT, the price of natura XL
xl_price = page.find_all('span', class_="product-price")
natura_xl_price = [item.text.strip() for item in xl_price]
print(natura_xl_price)
```

```python
# Now, we want to obatain the delivery time. Since the product page does not display any delivery time or info, we can rather indicate if the product is in stock or out of stock
xl_delivery = page.find_all('span', class_="stock-green")
natura_xl_delivery = [item.text.strip() for item in xl_delivery]
print(natura_xl_delivery)
```

```python
# Composition
xl_comp = page.find_all('tr')
natura_xl_comp = [item.text.strip() for item in xl_comp]
natura_xl_comp[8][11:] 
```

```python
# Needle 
xl_ndsize = page.find_all('tr')
natura_xl_ndsize = [item.text.strip() for item in xl_ndsize] 
natura_xl_ndsize[9][11:]
```
